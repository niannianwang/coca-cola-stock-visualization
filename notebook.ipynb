```python
import pandas as pd
```

```python
df = pd.read_csv("./static/Coca-Cola_stock_history.csv")
df
```

```python
def clean_text(text):
    if len(text) > 10:
        return text[:10].strip()
    return text.strip()
```

```python
df['Date'] = df['Date'].apply(clean_text)
```

```python
df.to_csv("./static/Coca-Cola_stock_history.csv", index=False)
```
